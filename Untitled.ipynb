```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
credito = pd.read_csv("salario_vs_sexo_vs_limite.csv")
```

```python
credito.info()
```

```python
sns.barplot(
    data = credito,
    x = "salario_anual",
    y =  "limite_medio_vs_salario",
    hue = "sexo",
    pallete = "M": "blue", "F":"pink"]
)
plt.xticks(rotation = 45)
plt.savefig("Grafico salario vs limite vs genero")
plt.show()
```
